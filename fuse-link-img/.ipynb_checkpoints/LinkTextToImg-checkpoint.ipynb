```python
# Cell 2: imports + config

import os, json, re
import numpy as np
import pandas as pd
from tqdm import tqdm
from PIL import Image, ImageFile
import torch
import faiss
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional

GEMINI_AVAILABLE = False

@dataclass
class Config:
    texts_dir: str = "./texts"
    images_dir: str = "./images"
    out_dir: str = "./out"

    clip_model_name: str = "openai/clip-vit-base-patch32"

    # indexing images
    faiss_image_index_path: str = "./out/clip_image.index"
    image_meta_path: str = "./out/image_meta.json"

    # assignment
    top_k_images: int = 25
    max_uses_per_image: int = 5
    lambda_penalty: float = 0.15

cfg = Config()
os.makedirs(cfg.out_dir, exist_ok=True)
```

```python
# Cell 3: helpers (IO + image filename parsing)

import os
import json
import re
import hashlib
from typing import Dict, Optional

def stable_id(s: str) -> str:
    return hashlib.sha1(s.encode("utf-8")).hexdigest()[:16]

def read_text_file(path: str) -> str:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()

def parse_image_filename(filename: str) -> Dict[str, Optional[str]]:
    """
    Expected filename format:
    A|P|I_Artist_Title_Date_(index).jpg
    """
    base = os.path.splitext(os.path.basename(filename))[0]
    parts = base.split("_")

    out = {
        "raw": base,
        "type": None,
        "artist": None,
        "title": None,
        "date": None,
        "series_index": None,
    }

    if len(parts) >= 2 and parts[0] in {"A", "P", "I"}:
        out["type"] = parts[0]
        tail = parts[1:]

        # optional series index (e.g. (1))
        if tail and re.fullmatch(r"\(?\d+\)?", tail[-1]):
            out["series_index"] = re.sub(r"[()]", "", tail[-1])
            tail = tail[:-1]

        # optional date
        if tail and re.fullmatch(r"\d{4}|\d{4}-\d{2}-\d{2}", tail[-1]):
            out["date"] = tail[-1]
            tail = tail[:-1]

        if len(tail) >= 2:
            out["artist"] = tail[0].strip()
            out["title"] = "_".join(tail[1:]).strip()
        elif len(tail) == 1:
            out["artist"] = tail[0].strip()

    return out

def load_json(path: str) -> Dict:
    if os.path.exists(path):
        with open(path, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def save_json(path: str, obj: Dict) -> None:
    with open(path, "w", encoding="utf-8") as f:
        json.dump(obj, f, ensure_ascii=False, indent=2)
```

```python
import re
from typing import List, Dict, Optional

# Main header: E3.9.4. Title...
MAIN_HEADER_RE = re.compile(r"^(?P<main_code>E\d+(?:\.[0-9A-Za-z]+)*)\.\s+(?P<title>.+?)\s*$")

# Pure code line: E3.9.13.a.
PURE_CODE_RE = re.compile(r"^(?P<code>E\d+(?:\.[0-9A-Za-z]+)*)\.\s*$")

# Internal EL header line: EL3.87.A.1. Something...
# It also matches EL3.87. (with nothing after)
EL_HEADER_RE = re.compile(r"^(?P<el>EL\d+(?:\.[0-9A-Za-z]+)*)\.\s*(?P<rest>.*)$")

def _normalise(text: str) -> str:
    return text.replace("\r\n", "\n").replace("\r", "\n")

def is_entry_header(line: str) -> bool:
    s = line.strip()
    if not s.startswith("E"):
        return False
    if PURE_CODE_RE.match(s):
        return True
    if "_REF" in s or re.search(r"_RE\d+_", s):
        return True
    if ".wE" in s:
        return True
    return False

def extract_chunk_code(line: str) -> str:
    s = line.strip()

    m_pure = PURE_CODE_RE.match(s)
    if m_pure:
        return m_pure.group("code")

    # E3.9.4_REF2.2_LIT ... or E2.2.A.3.wE3.9.4.
    return s.split()[0].rstrip(".")

def is_el_header(line: str) -> Optional[Dict[str, str]]:
    s = line.strip()
    m = EL_HEADER_RE.match(s)
    if not m:
        return None
    return {"el_code": m.group("el").strip(), "rest": (m.group("rest") or "").strip()}

def parse_text_to_chunks(raw: str, min_chars: int = 30) -> List[Dict[str, str]]:
    raw = _normalise(raw)
    lines = raw.split("\n")

    chunks: List[Dict[str, str]] = []

    current_section_code: Optional[str] = None
    current_section_title: str = ""

    current_chunk_code: Optional[str] = None
    buffer: List[str] = []

    # Para split interno por EL
    current_inner_code: Optional[str] = None

    def flush():
        nonlocal buffer, current_inner_code
        if current_section_code is None:
            buffer = []
            current_inner_code = None
            return

        text = "\n".join(buffer).strip()
        if len(text) < min_chars:
            buffer = []
            current_inner_code = None
            return

        # chunk_code final: se tiver inner EL, vira "E4..._REF2_PSI|EL3.87.A.1"
        final_code = current_chunk_code or current_section_code
        if current_inner_code:
            final_code = f"{final_code}|{current_inner_code}"

        chunks.append({
            "section_code": current_section_code,
            "section_title": current_section_title,
            "chunk_code": final_code,
            "chunk_text": text
        })

        buffer = []
        current_inner_code = None

    for line in lines:
        s = line.strip()

        # 1) New main section
        m = MAIN_HEADER_RE.match(s)
        if m:
            flush()
            current_section_code = m.group("main_code")
            current_section_title = m.group("title")
            current_chunk_code = None
            buffer = []
            continue

        if current_section_code is None:
            continue

        # 2) New chunk entry inside section (E..._REF / .wE / pure code)
        if is_entry_header(s):
            flush()
            current_chunk_code = extract_chunk_code(s)
            current_inner_code = None
            buffer = [line]
            continue

        # 3) Split inside chunk when EL header appears
        el = is_el_header(s)
        if el:
            # Se já existe conteúdo no buffer, fecha o chunk anterior e abre outro
            if buffer and "\n".join(buffer).strip():
                flush()
            current_inner_code = el["el_code"]
            buffer = [line]
            continue

        # Split extra: um novo header REF no meio do chunk (evita juntar dois REFs num registro)
        is_ref_header_line = s.startswith("E") and ("_REF" in s or re.search(r"_RE\d+_", s))
        if buffer and is_ref_header_line:
            already_has_ref = any(
                (b.strip().startswith("E") and ("_REF" in b or re.search(r"_RE\d+_", b)))
                for b in buffer
            )
            if already_has_ref:
                flush()
                buffer = [line]
                continue

        buffer.append(line)

    flush()
    return chunks
```

```python
#Cell 5: build do dataframe de chunks (chunks_df)

import os
import pandas as pd
import hashlib

def read_text_file(path: str) -> str:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()

def make_chunk_id(
    source_file: str,
    section_code: str,
    chunk_code: str,
    chunk_text: str) -> str:
    key = f"{source_file}||{section_code}||{chunk_code}||{chunk_text}"
    return hashlib.sha1(key.encode("utf-8")).hexdigest()


def build_chunks_df(texts_dir: str) -> pd.DataFrame:
    rows = []
    for root, _, files in os.walk(texts_dir):
        for fn in files:
            if not fn.lower().endswith((".txt", ".md")):
                continue
            path = os.path.join(root, fn)
            raw = read_text_file(path)

            chunks = parse_text_to_chunks(raw)
            for i, c in enumerate(chunks):
                chunk_id = make_chunk_id(
                    source_file=fn,
                    section_code=c["section_code"],
                    chunk_code=c["chunk_code"],
                    chunk_text=c["chunk_text"]
                )
            
                rows.append({
                    "chunk_id": chunk_id,          # <- NOVO ID ESTÁVEL
                    "source_file": fn,
                    "position": i,
                    "section_code": c["section_code"],
                    "section_title": c["section_title"],
                    "chunk_code": c["chunk_code"],
                    "chunk_text": c["chunk_text"],
                })


    df = pd.DataFrame(rows)
    if df.empty:
        raise RuntimeError("Nenhum chunk encontrado. Confira texts_dir e o formato dos arquivos.")
    return df.reset_index(drop=True)

chunks_df = build_chunks_df(cfg.texts_dir)
print("Chunks:", len(chunks_df))
chunks_df.head(5)

```

```python
# Cell 6: load CLIP

from transformers import CLIPModel, CLIPProcessor

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

clip_model = CLIPModel.from_pretrained(cfg.clip_model_name).to(device)
clip_processor = CLIPProcessor.from_pretrained(cfg.clip_model_name)

clip_model.eval()
print("Loaded:", cfg.clip_model_name)
```

```python
# Cell 7: CLIP text embeddings (no FAISS; texts will query an IMAGE index)

import os
import numpy as np
import torch
from typing import List
from tqdm import tqdm

@torch.no_grad()
def clip_text_embeddings(texts: List[str], batch_size: int = 64, max_len: int = 77) -> np.ndarray:
    vecs = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        inputs = clip_processor(
            text=batch,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_len
        )
        inputs = {k: v.to(device) for k, v in inputs.items()}
        feats = clip_model.get_text_features(**inputs)
        feats = feats / (feats.norm(dim=-1, keepdim=True) + 1e-12)  # normalise => IP == cosine
        vecs.append(feats.detach().cpu().numpy().astype(np.float32))
    return np.vstack(vecs)

# 1) Build chunk_meta (keep only what we need later)
required_cols = ["chunk_id", "chunk_code", "chunk_text"]
missing = [c for c in required_cols if c not in chunks_df.columns]
if missing:
    raise RuntimeError(f"chunks_df missing columns: {missing}. Check Cell 5 build_chunks_df.")

chunk_meta = chunks_df[required_cols].to_dict("records")

# 2) Text -> embeddings (we will query the IMAGE FAISS index with these vectors)
chunk_texts = chunks_df["chunk_text"].tolist()
chunk_vecs = clip_text_embeddings(chunk_texts, batch_size=64)

print("Computed chunk embeddings:", chunk_vecs.shape)

# 3) Save (optional but strongly recommended to avoid recomputing)
os.makedirs(cfg.out_dir, exist_ok=True)

chunk_vecs_path = os.path.join(cfg.out_dir, "chunk_vecs.npy")
chunk_meta_path = os.path.join(cfg.out_dir, "chunk_meta.json")

np.save(chunk_vecs_path, chunk_vecs)
save_json(chunk_meta_path, {"dim": int(chunk_vecs.shape[1]), "chunk_meta": chunk_meta})

print("Saved chunk_vecs:", chunk_vecs_path)
print("Saved chunk_meta:", chunk_meta_path)
```

```python
# Cell 8: Build IMAGE embeddings + FAISS image index (for text->image retrieval)

import os
import numpy as np
import faiss
import torch
from tqdm import tqdm
from typing import Dict, List, Tuple
from PIL import Image, ImageFile

# 1) Robust image loading settings
ImageFile.LOAD_TRUNCATED_IMAGES = True

# CLIP does not need gigantic images. Keep detail but avoid bombs.
MAX_SIDE = 2048  # if you want faster: 1024

@torch.no_grad()
def clip_image_embedding(image_path: str, max_side: int = MAX_SIDE) -> np.ndarray:
    """
    Returns shape [1, d] float32, L2-normalised.
    """
    img = Image.open(image_path)

    # Resize early to avoid huge decode/memory issues (best-effort).
    # Note: some images still trigger DecompressionBombError on open; we handle in the loop.
    w, h = img.size
    if max(w, h) > max_side:
        img = img.copy()
        img.thumbnail((max_side, max_side), Image.Resampling.LANCZOS)

    img = img.convert("RGB")

    inputs = clip_processor(images=img, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    feats = clip_model.get_image_features(**inputs)
    feats = feats / (feats.norm(dim=-1, keepdim=True) + 1e-12)
    return feats.detach().cpu().numpy().astype(np.float32)  # [1, d]

# 2) List images
image_files = sorted([
    fn for fn in os.listdir(cfg.images_dir)
    if fn.lower().endswith((".png", ".jpg", ".jpeg", ".webp"))
])
print("Images found:", len(image_files))

# 3) Compute embeddings + meta
image_vecs_list: List[np.ndarray] = []
image_meta: List[Dict[str, str]] = []
errors: List[Tuple[str, str, str]] = []

for fn in tqdm(image_files):
    image_path = os.path.join(cfg.images_dir, fn)
    try:
        vec = clip_image_embedding(image_path)  # [1, d]
        image_vecs_list.append(vec)
        meta = parse_image_filename(fn)  # you already have this helper
        meta_out = {
            "image_filename": fn,
            "type": meta.get("type") or "",
            "artist": meta.get("artist") or "",
            "title": meta.get("title") or "",
            "date": meta.get("date") or "",
            "series_index": meta.get("series_index") or "",
        }
        image_meta.append(meta_out)

    except Exception as e:
        errors.append((fn, type(e).__name__, str(e)))

if not image_vecs_list:
    raise RuntimeError("No image embeddings created. Check cfg.images_dir and image formats.")

image_vecs = np.vstack(image_vecs_list).astype(np.float32)  # [N, d]
dim = int(image_vecs.shape[1])
print("Image embeddings:", image_vecs.shape, "dim:", dim)

# 4) Build FAISS index for images (cosine via normalised inner product)
image_index = faiss.IndexFlatIP(dim)
image_index.add(image_vecs)

# 5) Save index + meta
os.makedirs(cfg.out_dir, exist_ok=True)

image_index_path = os.path.join(cfg.out_dir, "clip_image.index")
image_meta_path = os.path.join(cfg.out_dir, "image_meta.json")
errors_path = os.path.join(cfg.out_dir, "image_embedding_errors.json")

faiss.write_index(image_index, image_index_path)
save_json(image_meta_path, {"dim": dim, "image_meta": image_meta})
save_json(errors_path, {"errors": errors})

print("Saved image index:", image_index_path)
print("Saved image meta :", image_meta_path)
print("Image errors     :", errors_path, "count:", len(errors))
```

```python
# Cell 10: main loop (TEXT -> IMAGE) with repetition control
# Output: results_text_to_image.jsonl (one row per chunk)

import os
import json
import numpy as np
from tqdm import tqdm
import faiss

# Paths
os.makedirs(cfg.out_dir, exist_ok=True)
results_path = os.path.join(cfg.out_dir, "results_text_to_image.jsonl")

image_index_path = os.path.join(cfg.out_dir, "clip_image.index")
image_meta_path  = os.path.join(cfg.out_dir, "image_meta.json")

# ---------- Load IMAGE index + meta ----------
image_index = faiss.read_index(image_index_path)
image_blob = load_json(image_meta_path)
image_meta = image_blob["image_meta"]

def get_image(pid: int) -> dict:
    return image_meta[int(pid)]

# ---------- Load CHUNK meta + vectors ----------
# Option A: if you kept these in memory from Cell 7:
#   chunk_vecs: np.ndarray [n_chunks, d]
#   chunk_meta: List[dict] with chunk_id/chunk_code/chunk_text
#
# Option B: load from disk if you saved them:
# chunk_vecs = np.load(os.path.join(cfg.out_dir, "chunk_vecs.npy"))
# chunk_blob = load_json(os.path.join(cfg.out_dir, "chunk_meta.json"))
# chunk_meta = chunk_blob["chunk_meta"]

if "chunk_meta" not in globals():
    meta_blob = load_json(cfg.para_meta_path)  # where you saved chunk_meta in Cell 7
    if "chunk_meta" not in meta_blob:
        raise KeyError(f"Expected chunk_meta in {cfg.para_meta_path}. Keys: {list(meta_blob.keys())}")
    chunk_meta = meta_blob["chunk_meta"]

if "chunk_vecs" not in globals():
    # If you didn't save vectors yet, do it in Cell 7. For now, fail loudly.
    raise RuntimeError("chunk_vecs not found in memory. Save/load chunk_vecs.npy from Cell 7.")

n_chunks = len(chunk_meta)
n_images = len(image_meta)
print("Chunks:", n_chunks, "Images:", n_images)

# ---------- Repetition control ----------
# Max uses per image: choose a cap. Since 5270 texts / 2872 images ~ 1.83 uses/image average.
# A safe cap is 2 or 3. If you set 2, you'll still need some 3s because 2*2872=5744 covers 5270.
MAX_USES_PER_IMAGE = getattr(cfg, "max_uses_per_image", 2)

# For each chunk we will retrieve top_k images and pick the best that respects cap (best-effort).
TOP_K_IMAGES = getattr(cfg, "top_k_images", 30)

image_use_count = np.zeros(n_images, dtype=np.int32)

# Helper: load_done to resume
def load_done_chunks(path: str) -> set:
    done = set()
    if not os.path.exists(path):
        return done
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                obj = json.loads(line)
                cid = obj.get("chunk_id") or obj.get("chunk_code")
                if cid:
                    done.add(cid)
            except Exception:
                continue
    return done

done = load_done_chunks(results_path)
print("Already done chunks:", len(done))

# We also need to rebuild use_count when resuming
if done:
    with open(results_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                obj = json.loads(line)
            except Exception:
                continue
            img_pid = obj.get("image_pid", None)
            if isinstance(img_pid, int) and 0 <= img_pid < n_images:
                image_use_count[img_pid] += 1

# ---------- Search: for each chunk, retrieve candidate images ----------
# chunk_vecs already normalised in Cell 7, so IP==cosine
with open(results_path, "a", encoding="utf-8") as out_f:
    for i in tqdm(range(n_chunks)):
        c = chunk_meta[i]
        chunk_id = c.get("chunk_id", "")
        chunk_code = c.get("chunk_code", "")
        key = chunk_id or chunk_code
        if key in done:
            continue

        q = chunk_vecs[i:i+1].astype(np.float32)  # [1, d]

        scores, ids = image_index.search(q, TOP_K_IMAGES)
        scores = scores[0]
        ids = ids[0]

        # pick best image under cap; if none under cap, pick best anyway (soft constraint)
        chosen_pid = None
        chosen_score = None

        for pid, sc in zip(ids, scores):
            if pid < 0:
                continue
            if image_use_count[pid] < MAX_USES_PER_IMAGE:
                chosen_pid = int(pid)
                chosen_score = float(sc)
                break

        if chosen_pid is None:
            # all top-K are at cap; pick the best one anyway
            valid = [(int(pid), float(sc)) for pid, sc in zip(ids, scores) if int(pid) >= 0]
            if valid:
                chosen_pid, chosen_score = valid[0]
            else:
                chosen_pid, chosen_score = -1, 0.0

        if chosen_pid >= 0:
            image_use_count[chosen_pid] += 1
            im = get_image(chosen_pid)
        else:
            im = {"image_filename": "", "artist": "", "title": "", "date": "", "type": "", "series_index": ""}

        obj = {
            "chunk_id": chunk_id,
            "chunk_code": chunk_code,
            "image_pid": int(chosen_pid),
            "image_filename": im.get("image_filename", ""),
            "image_artist": im.get("artist", ""),
            "image_title": im.get("title", ""),
            "image_date": im.get("date", ""),
            "score": float(chosen_score or 0.0),
            "status": "ok" if chosen_pid >= 0 else "no_image",
            "top_candidates": [
                {"image_pid": int(pid), "score": float(sc),
                 "image_filename": image_meta[int(pid)]["image_filename"] if int(pid) >= 0 else ""}
                for pid, sc in zip(ids[:10], scores[:10]) if int(pid) >= 0
            ],
        }

        out_f.write(json.dumps(obj, ensure_ascii=False) + "\n")
        out_f.flush()

print("Saved:", results_path)
print("Images used at least once:", int((image_use_count > 0).sum()), "/", n_images)

# ---------- Post-fix: ensure every image used at least once ----------
# If some images are unused, assign them to chunks where swapping costs little:
unused = np.where(image_use_count == 0)[0].tolist()
print("Unused images:", len(unused))

if unused:
    # Load all current assignments into memory (one pass)
    rows = []
    with open(results_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                rows.append(json.loads(line))
            except Exception:
                pass

    # Build a map from chunk_code/id -> row index
    key_to_idx = {}
    for idx, r in enumerate(rows):
        k = r.get("chunk_id") or r.get("chunk_code")
        if k:
            key_to_idx[k] = idx

    # For each unused image, find a chunk where this image is in top candidates and the score gap is small.
    # Simple heuristic: re-search that chunk for this image and compare to current chosen score.
    for pid in tqdm(unused):
        # choose a target chunk with "room" (whose current image is overused)
        # we prefer chunks whose current chosen image has use_count > 1
        best_swap = None  # (delta, chunk_idx, new_score)

        for chunk_i in range(n_chunks):
            # quick skip: if row missing (shouldn't), continue
            c = chunk_meta[chunk_i]
            k = c.get("chunk_id") or c.get("chunk_code")
            if k not in key_to_idx:
                continue

            r = rows[key_to_idx[k]]
            cur_pid = r.get("image_pid", -1)
            cur_score = float(r.get("score", 0.0))

            # only consider swapping if current image is used more than 1 time (helps distribution)
            if cur_pid >= 0 and image_use_count[cur_pid] <= 1:
                continue

            q = chunk_vecs[chunk_i:chunk_i+1].astype(np.float32)
            sc, ids = image_index.search(q, TOP_K_IMAGES)
            sc = sc[0]; ids = ids[0]

            # check if pid appears
            found = None
            for p2, s2 in zip(ids, sc):
                if int(p2) == int(pid):
                    found = float(s2)
                    break
            if found is None:
                continue

            delta = cur_score - found  # how much we lose by swapping (smaller is better)
            if best_swap is None or delta < best_swap[0]:
                best_swap = (delta, key_to_idx[k], found, cur_pid)

        if best_swap is None:
            # fallback: just overwrite the lowest-confidence row
            # (this is rare if TOP_K is reasonable)
            lowest_idx = int(np.argmin([float(r.get("score", 0.0)) for r in rows]))
            cur_pid = rows[lowest_idx].get("image_pid", -1)
            best_swap = (999.0, lowest_idx, 0.0, cur_pid)

        _, row_idx, new_score, old_pid = best_swap

        # apply swap
        rows[row_idx]["image_pid"] = int(pid)
        im = get_image(pid)
        rows[row_idx]["image_filename"] = im.get("image_filename", "")
        rows[row_idx]["image_artist"] = im.get("artist", "")
        rows[row_idx]["image_title"] = im.get("title", "")
        rows[row_idx]["image_date"] = im.get("date", "")
        rows[row_idx]["score"] = float(new_score)
        rows[row_idx]["status"] = "forced_use_once"

        # update counters
        image_use_count[pid] += 1
        if isinstance(old_pid, int) and old_pid >= 0:
            image_use_count[old_pid] = max(0, image_use_count[old_pid] - 1)

    # rewrite file clean
    tmp_path = results_path + ".tmp"
    with open(tmp_path, "w", encoding="utf-8") as f:
        for r in rows:
            f.write(json.dumps(r, ensure_ascii=False) + "\n")
    os.replace(tmp_path, results_path)

    print("Post-fix done. Images used at least once:", int((image_use_count > 0).sum()), "/", n_images)
```

```python
# Cell 11: export final CSV (one row per CHUNK / text)

import json
import pandas as pd
import os

final_csv = os.path.join(cfg.out_dir, "final_pairings_text_to_image.csv")

rows = []
with open(results_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            obj = json.loads(line)
        except Exception:
            continue

        rows.append({
            "chunk_id": obj.get("chunk_id", ""),
            "chunk_code": obj.get("chunk_code", ""),
            "image_pid": obj.get("image_pid", -1),
            "image_filename": obj.get("image_filename", ""),
            "image_artist": obj.get("image_artist", ""),
            "image_title": obj.get("image_title", ""),
            "image_date": obj.get("image_date", ""),
            "score": obj.get("score", 0.0),
            "status": obj.get("status", ""),
        })

df = pd.DataFrame(rows)

# limpeza simples
df["score"] = pd.to_numeric(df["score"], errors="coerce").fillna(0.0)

# ordena por score desc (melhores associações no topo)
df = df.sort_values(["score", "chunk_code"], ascending=[False, True]).reset_index(drop=True)

df.to_csv(final_csv, index=False, encoding="utf-8")
print("Wrote:", final_csv)

df.head(10)
```
